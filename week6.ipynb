# 1. 와인 데이터

## 1. 와인 데이터 받기

```python
import pandas as pd

red_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master' +\
                                    '/dataset/winequality-red.csv'
white_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master' +\
                                    '/dataset/winequality-white.csv'

red_wine = pd.read_csv(red_url, sep=';')
white_wine = pd.read_csv(white_url, sep=';')
```

```python
red_wine.head()
```

```python
white_wine.head()
```

## 2. 데이터 정리하고 저장

```python
# 와인에 color column 추가
red_wine['color'] = 1
white_wine['color'] = 0

# DF 합치기
wine = pd.concat([red_wine, white_wine])
wine.info()
```

```python
# Quality column의 unique를 확인하여 와인 등급 체크
wine['quality'].unique()
```

### Quality Column의 히스토그램 확인

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.hist(wine['quality'], bins=6, facecolor='blue', alpha=0.5)
plt.show()
```

### 등급별 count 조회

```python
wine['quality'].value_counts()
```
